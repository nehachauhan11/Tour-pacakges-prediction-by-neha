```python
df = pd.get_dummies(df,drop_first = True)
```

```python
df.head()
```

```python
x = df.drop('NumberOfPersonVisiting', axis = 1)
y = df['NumberOfPersonVisiting']
```

```python
print(x)
```

```python
print(y)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2, random_state=42)
```

```python
# Feature Scalling
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
x_train = scaler.fit_transform(x_train)
x_train = scaler.transform(x_train)
```

```python
# Model Creation
from sklearn.linear_model import LogisticRegression
L_reg = LogisticRegression()
L_reg.fit(x_train, y_train)
```

```python
y_pred = L_reg.predict(x_train)
```

```python
y_pred
```

```python
# Model Accuracy/Score
```

```python
print('Simple Logistic Model Score:', L_reg.score(x_train, y_train)*100)
```

```python
# decision tree
```

```python
from sklearn.tree import DecisionTreeClassifier
model_dt=DecisionTreeClassifier()
model_dt.fit(x_train,y_train)
```

```python
## Predicting the  test result
```

```python
y_pred = model_dt.predict(x_test)
```

```python
print('Decisin Tree Model Score:',model_dt.score(x_train,y_train)*100)
```

```python
print('Decision Tree Test Accuracy:',model_dt.score(x_test,y_test)*100)
```

```python
## confusion matrix
```

```python
from sklearn.metrics import confusion_matrix,classification_report
```

```python
cm=confusion_matrix(y_pred,y_test)
cm
```
